 <H3>
Objective
</H3>
<P>
The goal of this course is to study the elements that support to student realize  the activities used for computer programming and logical mathematical . The student must know <b>basics mathematics, data structures, programming, differential equations, computer science, statistics, linear algebra, probability, discrete mathematics and one variable calculus</b>.

The system engineer must be capable to estimates matematically the problem's computational complexity implicit in professional activity taking into account the performance (time response  and space) of his implemented solution.


</P>
