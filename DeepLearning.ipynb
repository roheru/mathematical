 <H3>
Objective
</H3>
<P>
The goal of this course is to study the elements that support to student realize  the activities used for mathematical and mathematical modelling, applied to real life. The student must <b>know basics mathematics, stochastic processes, computer programming, differential equations, computer science, statistics, linear algebra, probability, discrete mathematics, multi-variable calculus and operations research</b>.
</P>
